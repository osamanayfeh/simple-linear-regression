```python
## Library 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
import seaborn as sns # this is used for the plot the graph
import matplotlib.pyplot as plt # this is used for the plot the graph
from sklearn.model_selection import train_test_split # to split the data into two parts
from sklearn.linear_model import LinearRegression # to apply the Linear regression model
from sklearn.metrics import r2_score, mean_squared_error # for checking the model accuracy
import os # to use operating system dependent functionality
```

```python
data = pd.read_csv('salary_data.csv') # load the data set
```

```python
data.head() # display the first five rows of the data set
```

```python
data.info() # display the data information
```

```python
data.describe() # display the data description
```

```python
sns.pairplot(data) # plot the pair plot
```

```python
X = data.iloc[:, :-1] # X is the feature set
y = data.iloc[:, 1] # y is the target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10) # split the data into training and test set
```

```python
My_Model = LinearRegression() # create a model object
My_Model.fit(X_train, y_train) # fit the model object
print(f"test the Train_model: {My_Model.score(X_train, y_train)}") # test the model
print(f"test the Test_model: {My_Model.score(X_test, y_test)}") # test the model
```

**Plooting the training set**

```python
plt .scatter(X_train, y_train, color = 'red') # plotting the training set
plt.plot(X_train, My_Model.predict(X_train), color = 'blue') # plotting the line made by the model
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

**Plooting the testing set**

```python
y_pred = My_Model.predict(X_test) # make predictions
plt.scatter(X_test, y_test, color = 'red') # plotting the test set
plt.plot(X_test, y_pred, color = 'blue') # plotting the line made by the model
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
